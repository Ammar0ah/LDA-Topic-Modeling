```python
import numpy as np
```

```python
def lower_case(data):
    return np.char.lower(data)
```

```python

def remove_stop_words(data):
    stopword = stopwords.words("English")
    words = word_tokenize(str(data))
    new_text = ""
    for w in words:
        if w not in stopword and len(w) > 1:
            new_text += w + " "
    return new_text
```

```python
def remove_punctuation(data):
    symbols = "!\"#$%&()*+-./:;'<=>?@[\]^_`{|}~\n"
    for i in range(len(symbols)):
        data = np.char.replace(data, symbols[i], '')
#         data = np.char.replace(data, "  ", " ")
    data = np.char.replace(data, ',', '')
    return data
```

```python
def remove_apostrophe(data):
    return np.char.replace(data , "'"," ")
```

```python
def lemmitizing(data):
    lemmitizer = WordNetLemmatizer()
    tokens = word_tokenize(str(data))
    lem = ""
    for w in tokens:
        lem += " " + lemmitizer.lemmatize(w)
    return lem
```

```python

def preprocess(data):
    data = lower_case(data)
#     data = remove_apostrophe(data)
    data = remove_stop_words(data)
    data = lemmitizing(data) #needed again as we need to stem the words
    data = remove_punctuation(data) #needed again as num2word is giving few hypens and commas fourty-one
    data = remove_stop_words(data) #needed again as num2word is giving stop words 101 - one hundred and one
    return data
```

```python
stopwords = nltk.corpus.stopwords.words("english")
def clean_data(data):
    data = "".join([word for word in data if word not in string.punctuation])
    data = re.split("\W+",data)
    data = " ".join([word for word in data if word not in stopwords])
    return data
```
